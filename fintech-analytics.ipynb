```python
import pandas as pd
data = pd.read_csv("C:/Users/Kabir Verma/OneDrive/Desktop/New folder/transactions.csv")

class transaction:
    def __init__(self, txn_id, customr_id, merchant_id, amount, timestamp, channel, status):
        self.amount = float(amount)
        self.channel = channel
        self.txn_id = txn_id
        self.customr_id = customr_id
        self.merchant_id = merchant_id
        self.status = status
        self.timestamp = timestamp

    def stat(self):
        if self.status == "SUCCESS":
            print("Transaction successful")
        else:
            print("Transaction failed")

    def chnl(self):
        if self.channel == "Wallet":
            print("Wallet channel")
        elif self.channel == "UPI":
            print("UPI channel")
        elif self.channel == "Card":
            print("Card channel")
        elif self.channel == "QR":
            print("QR channel")
        else:
            print("Channel not found")

o1 = transaction("TXN286721", "CUST0442", "MERCH084", 4500.17, "2024-03-28", "UPI", "SUCCESS")
o2 = transaction("TXN364140",    "CUST0683",    "MERCH119",   123.12,  "2024-04-03",  "Wallet",  "SUCCESS")
o3 = transaction("TXN923483",    "CUST0873",    "MERCH160",  2965.06,  "2024-03-25",     "UPI",   "FAILED")
o4 = transaction("TXN522547",    "CUST0547",    "MERCH147",  1317.70,  "2024-06-22",    "Card",   "FAILED")
o5 = transaction("TXN869801",    "CUST0540",    "MERCH189",  2899.57,  "2024-03-14",      "QR",  "SUCCESS")
o1.stat()
o1.chnl()
print(" ")
o2.stat()
o2.chnl()
print(" ")
o3.stat()
o3.chnl()
print(" ")
o4.stat()
o4.chnl()
print(" ")
o5.stat()
o5.chnl()
```

```python
import numpy as np
a1 = np.array(data.amount)
print(a1)
def bbsort(a1):
    n = len(a1)
    for i in range(n):
        for j in range(0, n-i-1):
            if a1[i] < a1[j]:
                c = a1[j]
                a1[j] = a1[j+1]
                a1[j+1] = c

bbsort(a1)
print(" ")
print(a1)
```
